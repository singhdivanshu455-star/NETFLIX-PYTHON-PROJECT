# 📈 Data Visualization

## Overview

Data visualization helps us understand patterns, trends, and relationships within the Netflix dataset. The following charts provide insights into Netflix's content library, including content types, countries, genres, ratings, release years, and more.

## 🎬 1. Movies vs TV Shows
This chart compares the number of Movies and TV Shows available on Netflix.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=NETFLIX,
              x='type',
              hue='type',
              palette=[NETFLIX_RED, '#6D6D6D'],
              legend=False)
plt.title("Movies vs TV Shows", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.savefig("movies_vs_tvshows.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

## 🥧 2. Movies vs TV Shows Percentage
This pie chart shows the percentage distribution of Movies and TV Shows.

```python
content = NETFLIX['type'].value_counts()
plt.figure(figsize=(7,7), facecolor='black')
plt.pie(content, labels=content.index, autopct='%1.1f%%', startangle=90,
        explode=(0.05,0), colors=[NETFLIX_RED, '#6D6D6D'],
        textprops={'color':'white'})
plt.title("Movies vs TV Shows (%)", fontsize=16, color='white')
plt.savefig("movies_vs_tvshows_pie.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

## 🌍 3. Top 10 Countries by Content
This chart displays the countries with the highest number of Netflix titles.

```python
top_country = NETFLIX['country'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_country.values,
            y=top_country.index,
            color=NETFLIX_RED)
plt.title("Top 10 Countries", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("top_10_countries.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

## ⭐ 4. Rating Distribution
This chart shows the distribution of content ratings on Netflix.

```python
plt.figure(figsize=(10,6))
sns.countplot(y='rating', data=NETFLIX,
              order=NETFLIX['rating'].value_counts().index,
              color=NETFLIX_RED)
plt.title("Content Rating Distribution", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Count")
plt.ylabel("Rating")
plt.tight_layout()
plt.savefig("content_rating_distribution.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

## 📅 5. Release Year Distribution
This histogram shows how Netflix content is distributed by release year.

```python
plt.figure(figsize=(12,6))
sns.histplot(NETFLIX['release_year'], bins=30, kde=True, color=NETFLIX_RED)
plt.title("Content Released by Year (Distribution)", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("release_year_distribution.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Key Insights

- Movies represent a larger share of Netflix's catalog than TV Shows.
- The United States contributes the highest number of titles.
- TV-MA is one of the most common content ratings.
- Netflix expanded its content library significantly after 2015.

## 🎭 6. Top 10 Genres

This chart displays the most popular genres available on Netflix.

```python
top_genres = NETFLIX['listed_in'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(
    x=top_genres.values,
    y=top_genres.index,
    color=NETFLIX_RED
)
plt.title("Top 10 Genres on Netflix", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.savefig("top_10_genres.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

Drama and International Movies are among the most popular genres on Netflix, showing a strong focus on diverse global entertainment.

## 🎬 7. Top 10 Directors

This chart highlights the directors with the highest number of titles on Netflix.

```python
top_directors = NETFLIX['director'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(
    x=top_directors.values,
    y=top_directors.index,
    color=NETFLIX_RED
)
plt.title("Top 10 Directors on Netflix", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.tight_layout()
plt.savefig("top_10_directors.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

A small number of directors have contributed multiple titles, indicating long-term collaborations with Netflix.

```python
release = NETFLIX['release_year'].value_counts().sort_index()
plt.figure(figsize=(14,6))
sns.lineplot(x=release.index, y=release.values, marker='o', color=NETFLIX_RED)
plt.title("Content Released by Year", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.savefig("release_year_trend.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

Netflix experienced rapid content growth after 2015, reflecting significant investment in expanding its global library.

```python
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']
plt.figure(figsize=(12,6))
sns.countplot(data=NETFLIX, x='month_added', order=month_order, color=NETFLIX_RED)
plt.xticks(rotation=45)
plt.title("Content Added by Month", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Month")
plt.ylabel("Titles Added")
plt.tight_layout()
plt.savefig("content_added_by_month.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

Content additions vary throughout the year, with certain months showing noticeably higher activity.

## ⏱️ 10. Movie Duration Distribution

This histogram shows the distribution of movie durations available on Netflix.

```python
movies = NETFLIX[NETFLIX['type'] == 'Movie'].copy()
movies['duration'] = movies['duration'].str.replace(' min', '', regex=False)
movies['duration'] = pd.to_numeric(movies['duration'], errors='coerce')

print(movies[['title', 'duration']].head())
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=movies['duration'], color=NETFLIX_RED)
plt.title("Movie Duration Box Plot", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Duration (Minutes)")
plt.tight_layout()
plt.savefig("movie_duration_boxplot.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(movies['duration'].dropna(), bins=20, kde=True, color=NETFLIX_RED)
plt.title("Movie Duration Distribution", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.savefig("movie_duration_distribution.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

Most Netflix movies have a duration between 80 and 120 minutes, indicating a preference for standard feature-length films.

## 📦 11. Movie Duration Box Plot

This box plot helps identify the spread of movie durations and highlights potential outliers.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=movies['duration'], color=NETFLIX_RED)
plt.title("Movie Duration Box Plot", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Duration (Minutes)")
plt.tight_layout()
plt.savefig("movie_duration_boxplot.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

Most Netflix movies have a duration between **80 and 120 minutes**, while a few movies are significantly longer, appearing as outliers.

## 🌍 13. Top 15 Countries by Number of Titles

This chart identifies the countries contributing the highest number of titles on Netflix.

```python
country = NETFLIX['country'].dropna().str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=country.values, y=country.index, color=NETFLIX_RED)
plt.title("Top 10 Content Producing Countries", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("top_10_countries_exploded.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

The **United States** dominates Netflix's content library, followed by several countries contributing regional and international content.

```python
plt.figure(figsize=(12,6))
sns.countplot(data=NETFLIX, x='rating', hue='type',
              order=NETFLIX['rating'].value_counts().index,
              palette=[NETFLIX_RED, '#6D6D6D'])
plt.xticks(rotation=45)
plt.title("Rating Distribution by Content Type", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.savefig("rating_by_type.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Insight

Movies dominate most rating categories, while TV Shows are concentrated in fewer rating classifications.

## 🎭 16. Top 10 Actors on Netflix

This visualization shows the actors who have appeared in the highest number of Netflix titles.

```python
actors = (
    NETFLIX['cast']
    .dropna()
    .str.split(', ')
    .explode()
    .value_counts()
    .head(10)
)
plt.figure(figsize=(12,6))
sns.barplot(
    x=actors.values,
    y=actors.index,
    color=NETFLIX_RED
)
plt.title("Top 10 Actors on Netflix", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Number of Titles")
plt.ylabel("Actor")
plt.tight_layout()
plt.savefig("top_10_actors.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```

### 💡 Business Insight

A small group of actors appears frequently across Netflix titles, highlighting recurring collaborations.

```python
country = (
    NETFLIX['country']
    .dropna()
    .str.split(', ')
    .explode()
    .value_counts()
    .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=country.values,
    y=country.index
)

plt.title("Top 10 Content Producing Countries", fontsize=16, fontweight='bold')
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

```python
genre = NETFLIX['listed_in'].dropna().str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=genre.values, y=genre.index, color=NETFLIX_RED)
plt.title("Top 10 Netflix Genres", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.savefig("top_10_genres_exploded.png", dpi=300, bbox_inches='tight', facecolor='black')
plt.show()
```
